```python
import pandas as pd
import numpy as np

def parse_tags_array(tags_array):
    tag_dict = {}
    if isinstance(tags_array, np.ndarray):
        for tag in tags_array:
            if isinstance(tag, dict) and "key" in tag and "value" in tag:
                tag_dict[tag["key"]] = tag["value"]
    return tag_dict

for i in range(24):
    trace = f"trace_jaeger-span_2025-06-06_{i:02d}-00-00.parquet"
    df = pd.read_parquet("../phaseone/2025-06-06/trace-parquet/" + trace)
    print(f"Processing {trace} with {len(df)} records")
    traces = df.groupby('traceID')
    for trace_id, spans in traces:
        print(f"Trace ID: {trace_id}")
        for index, row in spans.iterrows():
            tags = parse_tags_array(row['tags'])
            # if tags.get("span.kind", "") == "client":
            trace_id = row['traceID']
            span_id = row['spanID']
            operation_name = row['operationName']
            references = row['references']
            start_time = pd.to_datetime(row["startTime"], unit="ns")
            duration = row['duration']
            tags = row['tags']
            logs = row['logs']
            process = row['process']
            print(f"traceID\t{trace_id}\nspanID\t{span_id}\noperationName\t{operation_name}\nreferences\t{references}\nstart_time\t{start_time}\nduration\t{duration}\ntags\t{tags}\nlogs\t{logs}\nprocess\t{process}")
        break
    break
```
